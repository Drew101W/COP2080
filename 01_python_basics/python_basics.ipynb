```python
# Declaring Variables with Type Annotations
name: str = "Drew"
age : int = 20
height: float = 6.2

print(f"Name: {name}, Age: {age}, Height: {height}")
```

```python
# Arithmetic
a = 5
b = 9
print(b / a)
print(b // a) # integer division
print(b**a) # exponent
print(b % a) # the remainder
```
