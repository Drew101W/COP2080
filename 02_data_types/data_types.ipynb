```python
#int function
print(int(35.4))
print(int("1000", 2))
print(int("100", 8)) 
```

```python
# bool function (True, False)
print(bool(True)) #True
print(bool(0)) #False
print(bool(-1)) #True
print(bool(not True)) #False
print(bool(True and True)) #True
print(bool(True or False)) #True
```

```python
# decimal
from decimal import Decimal
d = Decimal("0.1") + Decimal("0.1") + Decimal("0.1")
print(d)
```
