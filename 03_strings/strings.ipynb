```python
# string slicing 
# specify the start and end of the slice

language = "Pytnon"
language = language.replace("n", "h", 1)

print(language[0:3])
print(language)

#what is the output
print(language[2:len(language)])
print(language[2:])
print(language[:5])

```

```python
# you can use negative indices
language = "Python"

print(language[-3:])

# specify the step to
print(language[::2])
print(language[::-2])
# reverse a string
print(language[::-1])
```

```python
makes = "toyota, bmw, audi, subaru, nissan"

car_makes = makes.split(",") # car_makes is a list of strings
#interate through the list 
for make in car_makes:
    make = make.lstrip()
    if make == "bmw" or make == "audi":
        print(make.upper())
    else:
        print(make.title())    
```
